```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Dynamics function
def q(x, u, v):
    return  x - u - v   # State at the next time step, given current state and control input
```

```python
numEnsembles = 10**6 # how many ensembles will we run
nSteps = 20  # Number of steps to simulate for in each ensemble
x0 = 1    # Initial state

# Time step array
tArray = np.arange(nSteps+1) #same for all ensembles

# State history array
xArray = np.zeros((nSteps+1,numEnsembles))
fractionArray=np.zeros((nSteps+1,numEnsembles))
fractionstorage=np.zeros((nSteps+1,numEnsembles))
counter = 0

for ensemble in range(numEnsembles):
    xArray[0, ensemble] = x0    # Initializing the state
    
    #run the dynamics:
    for k in range(nSteps):
        u = 0.1
        v = np.random.normal(0,0.05)
        xArray[k+1, ensemble] = q(xArray[k, ensemble], u, v)
        
        #if (xArray[k+1, ensemble])<=0:
            
            #fractionArray[k+1,ensemble] = (counter/numEnsembles)
            #counter=counter+1
            
        if (xArray[k+1, ensemble])<=0:
            fractionstorage[k+1, ensemble]=1
            fractionArray[k+1,ensemble]=(np.sum(fractionstorage))/nSteps
            
```

We have the state history. Now we can plot it vs. time for visualization.

```python
# Plot the distribution and histogram
fig, ax = plt.subplots(1, 1)
ax.plot(tArray, xArray, label='State evolution')
ax.set_xlim([tArray[0], tArray[-1]])
ax.set_xlabel('Time step k')
ax.set_ylabel('State x(k)')

plt.show()
```

### Plot the ensemble statistics:
mean and std deviation at each time step:

```python
mean_xArray = np.mean(xArray,1)

mean_frac=np.mean(fractionstorage,1)

variance_xArray = np.var(xArray,1)

stnd_xArray=np.sqrt(variance_xArray)

#mean plot
fig, ax = plt.subplots(1, 1)
ax.plot(tArray, mean_xArray, label='E[q(k)]')
ax.set_ylabel('Mean q(k)')
ax.set_xlabel('Time step k')
ax.grid(True)

#variance plot
fig, ax = plt.subplots(1, 1, figsize=(6, 4))
ax.plot(tArray, variance_xArray, label='Var[q(k)]')
ax.set_ylabel('Variance q(k)')
ax.set_xlabel('Time step k')
ax.grid(True)

#standard deviation plot
fig, ax = plt.subplots(1, 1)
ax.plot(tArray, stnd_xArray, label='Stnd[q(k)]')
ax.set_ylabel('standard deviation q(x)')
ax.set_xlabel('Time step k')
ax.grid(True)


#fraction of sample paths plot
fig, ax=plt.subplots(1,1)
ax.plot(tArray, mean_frac, label='fraction[(q(k)<=0)/k]')
ax.grid(True)
ax.set_ylabel('fraction time')
ax.set_xlabel('Time step k')

plt.show()

print(mean_xArray[:,])
print('\n')
print(stnd_xArray[:,])
print('\n')
print(mean_frac[:,])
```

In a single graph, we can plot the sample mean, plus/minus one standard deviation. 

Recall -- if $x$ is normally distributed, then around 68% of samples should lie within one standard deviation (and 99.7% within 3 standard deviations)


```python
mean_xArray = np.mean(xArray,1)
variance_xArray = np.var(xArray,1)

fig, ax = plt.subplots(1, 1)
ax.plot(tArray, mean_xArray, label='E[x(k)]')
ax.plot(tArray, mean_xArray+3*np.sqrt(variance_xArray), 'r', label='E[x(k)]+std')
ax.plot(tArray, mean_xArray-3*np.sqrt(variance_xArray), 'r', label='E[x(k)]+std')
ax.set_ylabel('State q(k)')
ax.set_xlabel('Time step k')
ax.grid(True)

plt.show()
```
