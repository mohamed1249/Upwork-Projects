```python
print("Question: 1")
print("Part (A)")
print("Confirm that this function is a valid PDF")

import numpy as np
import sympy

t = sympy.symbols('t')
thousand = 1000
expy = sympy.exp(-t**2 / thousand**2)
ft_function = (2 * t / thousand**2) * expy
ft_function
```

```python
print("if the integration of the function is 1, means that its valid PDF")
sympy.integrate(ft_function, [t, 0 ,sympy.oo])
```

```python
print("Question: 3")
print("Part (E)")
import matplotlib.pyplot as plt
print("Define Dynamics function")
def q(x, u, v):
    return  x - u - v 
```

```python
print("The Code is Taking so much time therefore I took 10^4 value")
num_of_Ensembles, nSteps, x0 = 10**4, 20, 1
tArray, xArray = np.arange(nSteps+1), np.zeros((nSteps+1, num_of_Ensembles))
fractionArray, fractionstorage = np.zeros((nSteps+1, num_of_Ensembles)), np.zeros((nSteps+1, num_of_Ensembles))
counter = 0
for ensemble in range(num_of_Ensembles):
    xArray[0, ensemble] = x0  
    for k in range(nSteps):
        xArray[k+1, ensemble] = q(xArray[k, ensemble], 0.1, np.random.normal(0,0.05))    
        if (xArray[k+1, ensemble]) <= 0:
            fractionstorage[k+1, ensemble] = 1
            fractionArray[k+1,ensemble] = (np.sum(fractionstorage)) / nSteps        
```

```python
print("Plot State History and Time Plotting")
fig, ax = plt.subplots(1, 1, figsize=(12, 5))
ax.plot(tArray, xArray, label='Evolution of State')
ax.set_xlim([tArray[0], tArray[-1]])
ax.set_xlabel('Time Steps --> (k)')
ax.set_ylabel('State --> x(k)')
plt.grid()
plt.show()
```

```python
mean_xArray = np.mean(xArray, 1)
fig, ax = plt.subplots(1, 1)
ax.plot(tArray, mean_xArray, label='E[q(k)]')
ax.set_ylabel('Mean --> q(k)')
ax.set_xlabel('Time step --> k')
ax.set_title("Mean/ Average Plot")
ax.grid(True)
```

```python
mean_frac = np.mean(fractionstorage, 1)
fig, ax=plt.subplots(1,1)
ax.plot(tArray, mean_frac, label='fraction[(q(k) <= 0)/k]')
ax.set_ylabel('Fraction Time')
ax.set_xlabel('Time Step --> k')
ax.set_title("Fraction of Sample Paths Plot")
ax.grid(True)
plt.show()
```

```python
variance_xArray = np.var(xArray, 1)
fig, ax = plt.subplots(1, 1, figsize=(6, 4))
ax.plot(tArray, variance_xArray, label='Var[q(k)]')
ax.set_ylabel('Variance --> q(k)')
ax.set_xlabel('Time step --> k')
ax.set_title("Variance Plot")
ax.grid(True)
```

```python
stnd_xArray = np.sqrt(variance_xArray)
fig, ax = plt.subplots(1, 1)
ax.plot(tArray, stnd_xArray, label='Stnd[q(k)]')
ax.set_ylabel('Standard Deviation -> q(x)')
ax.set_xlabel('Time Step -> k')
ax.set_title("Standard Deviation Plot")
ax.grid(True)
```

```python
print("Question: 4")
print("Define Dyanmics, Noise Variance, Measurement Model")
A = np.matrix([[1]])
V = np.matrix([[9]])
H = np.matrix([[1]])
W = np.matrix([[25]])
```

```python
measurements_z = [np.nan, 32.1 , 39.8, 45.9, 52.0, 51.0, 63.4, 58.1, 60.3, 76.6, 73.0]
measurements_z
```

```python
inputs_uf = [10, 10, 10, 10, 10, 10, 10, 10, 10, 10, 10]
inputs_uf
```

```python
m = [7, 7, 7, 7, 7, 7, 7, 7, 7, 7, 7]
m
```

```python
hat_x_m = np.mat([[20]]) 
hat_x_m
```

```python
P_m = np.mat([[25]])
P_m
```

```python
numSteps = 10 
hist_hat_x_m, hist_P_m_diag = np.zeros([numSteps + 1]), np.zeros([numSteps + 1])  
hist_hat_x_m[0], hist_P_m_diag[0] = hat_x_m[0], P_m[0]
for k in np.arange(1,numSteps+1):
    uf = inputs_uf[k-1] - m[k-1]
    u = np.matrix([[uf]])
    hat_x_p = A*hat_x_m + u
    P_p = A**2*P_m + V
    z = measurements_z[k]
    K = P_p @ H.T @ np.linalg.inv(H @ P_p @ H.T + W)
    hat_x_m = hat_x_p + K @ (z - H @ hat_x_p)
    P_m = (np.eye(1) - K @ H) @ P_p @ (np.eye(1) - K @ H).T + K @ W @ K.T
    hist_hat_x_m[k] = hat_x_m[0]
    hist_P_m_diag[k] = P_m[0]
```

```python
fig, ax = plt.subplots(1,1,sharex=True, figsize=(10, 5))
ax.plot(hist_hat_x_m[:],'r.-',label="Estimated State")
ax.plot(hist_hat_x_m[:]+np.sqrt(hist_P_m_diag[:]),'g:',label="1 Sigma")
ax.plot(hist_hat_x_m[:]-np.sqrt(hist_P_m_diag[:]),'g:',)
ax.plot(measurements_z,'b*',label="Measerments z(k)")
ax.set_ylabel('r(k)')
ax.set_xlabel('Time Step -> k')
ax.legend()
ax.grid(True)
```

```python
fig, ax = plt.subplots(1,1,sharex=True, figsize=(10, 5))
ax.plot(hist_P_m_diag[:],'r*-',label="P(k)")
ax.set_xlabel('Time Step -> k')
ax.set_ylabel('P(k)')
ax.legend()
ax.grid(True)
```

```python
print("Now Solving EFK without measurements")
def get_A(x):
    return 1
def get_L(x):
    return -1
def get_H(x):
    return 3*(x-1)**2
def get_M(x):
    return 1
#Define noise variances
V = 0.05**2
W = 0.1**2
j0=0.1
```

```python
numSteps = 9
hat_x_m = 1
P_m = 0
hist_hat_x_m = np.zeros([numSteps+1,1])  
hist_P_m = np.zeros([numSteps+1,1]) 
hist_true_x = np.zeros([numSteps+1,1])  
true_state_x = np.random.normal(1,0)
hist_z = np.zeros([numSteps+1,1])
hist_hat_x_m[0,0] = hat_x_m
hist_P_m[0,0] = P_m
hist_true_x[0,0] = true_state_x
for k in np.arange(1,numSteps+1):
   #process noise
    v = np.random.normal(0, 0.05)
    #simulate the rocket:
    new_x = true_state_x - j0 - v
    true_state_x = new_x
    #generate the measurement:
    w = np.random.normal(0, 0.1)
    z = 4 + ((true_state_x)-1)**3 + w
    #z=measurment_z[k-1]
    
    hist_z[k,0]=z
    hat_x_p = hat_x_m - j0 - 0 
    A = get_A(hat_x_p)
    L = get_L(hat_x_p)
    P_p = A*P_m*A + L*V*L 
    
    #step 2: Kalman filter measurement update:
    H = get_H(hat_x_p)
    M = get_M(hat_x_p)
    K = P_p*H*1/(H*P_p*H + W) 
    
    predictedMeas = 4 + ((hat_x_p)-1)**3 + 0 
    hat_x_m = hat_x_p + K*(z-predictedMeas)
    I = 1
    P_m = (I - K*H)*P_p*(I - K*H) + K*W*K # (no transpose, all scalars)
    ## EKF END
    #store the variables for plotting. Don't modify the below lines:
    hist_hat_x_m[k,0] = hat_x_m
    hist_P_m[k,0] = P_m
    hist_true_x[k,0] = true_state_x
```

```python
fig, ax = plt.subplots(1,1,sharex=True, figsize=(10, 5))
ax.plot(hist_hat_x_m[:,0],'g*-',label="est")
ax.set_xlim(0,9)
ax.set_xlabel('Time k')
ax.set_ylabel('State estimated state of charge')
ax.legend()
ax.grid(True)
plt.show()
```

```python
print("Question: 5")
print("Part (C)")
def q(x, u, v):
    return  x - u - v  
nSteps = 9
x0 = 1
#constants
pp=0.1
W=0.1
# Time step array
tArray = np.arange(nSteps+1) #same for all ensembles
# State history array
xArray = np.zeros(nSteps+1)
hArray = np.zeros(nSteps+1)
uArray = np.zeros(nSteps+1)
a=np.zeros(nSteps+1)
b=np.zeros(nSteps+1)
xArray[0]=x0
hArray[0]=(xArray[0]-1)**2
a[0]=hArray[0]**2*pp
b[0]=hArray[0]**2*pp+W
uArray[0]=a[0]/b[0]
for k in range(nSteps):
        u = 0.1
        v = np.random.normal(0,0.05)
        xArray[k+1] = q(xArray[k], u, v)
        hArray[k+1] = 3*(xArray[k+1]-1)**2
        a[k+1]=pp*hArray[k+1]**2
        b[k+1]=pp*hArray[k+1]**2+W
        uArray[k+1] = a[k+1]/b[k+1]
```
