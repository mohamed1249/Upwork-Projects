## case when t_m = 1

```python
import math
import numpy as np

no_of_Samples=10**6 
tSamples=np.zeros((no_of_Samples,)) 
Runcost=np.zeros((no_of_Samples,))
t_m=1
cr=250
cm=50
for k in range(no_of_Samples):   
    u=np.random.uniform()
    tSamples[k]=(1000*math.log(-1/(u-1)))
    if tSamples[k]<=t_m:
        Runcost[k]=cr/tSamples[k]
    else:
        Runcost[k]=cm/t_m
sampleMean=np.mean(tSamples)
sampleVariance=np.var(tSamples)
avg_R=np.mean(Runcost)

print('Over', no_of_Samples,'samples:')
print('The sample mean is:', sampleMean)
print('The sample variance is:', sampleVariance)
print('Average cost for tm =', t_m,'is $'+str(round(avg_R,2)))
```

```python
import matplotlib.pyplot as plt
```

```python
# Dynamics function
def q(x, u, v):
    return  x - u - v 
```

```python
print("Taking a lot of time, hence change numEnsemble to 10^4")
numEnsembles = 10**4
nSteps = 20
x0 = 1
tArray = np.arange(nSteps+1)
xArray = np.zeros((nSteps+1,numEnsembles))
fractionArray = np.zeros((nSteps+1,numEnsembles))
fractionstorage = np.zeros((nSteps+1,numEnsembles))
counter = 0
for ensemble in range(numEnsembles):
    xArray[0, ensemble] = x0  
    for k in range(nSteps):
        u = 0.1
        v = np.random.normal(0,0.05)
        xArray[k+1, ensemble] = q(xArray[k, ensemble], u, v)    
        if (xArray[k+1, ensemble]) <= 0:
            fractionstorage[k+1, ensemble] = 1
            fractionArray[k+1,ensemble] = (np.sum(fractionstorage)) / nSteps        
```

## Plot State History vs Time

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 6))
ax.plot(tArray, xArray, label='State Evolution')
ax.set_xlim([tArray[0], tArray[-1]])
ax.set_xlabel('Time Steps (k)')
ax.set_ylabel('State x(k)')
plt.show()
```

<span style="color:blue">

```python
mean_xArray = np.mean(xArray, 1)
mean_frac = np.mean(fractionstorage, 1)
variance_xArray = np.var(xArray, 1)
stnd_xArray = np.sqrt(variance_xArray)
mean_xArray, mean_frac, variance_xArray, stnd_xArray
```

```python
print("Mean Plot")
fig, ax = plt.subplots(1, 1)
ax.plot(tArray, mean_xArray, label='E[q(k)]')
ax.set_ylabel('Mean q(k)')
ax.set_xlabel('Time step k')
ax.grid(True)
```

```python
print("Variance Plot")
fig, ax = plt.subplots(1, 1, figsize=(6, 4))
ax.plot(tArray, variance_xArray, label='Var[q(k)]')
ax.set_ylabel('Variance q(k)')
ax.set_xlabel('Time step k')
ax.grid(True)
```

```python
print("Standard Deviation Plot")
fig, ax = plt.subplots(1, 1)
ax.plot(tArray, stnd_xArray, label='Stnd[q(k)]')
ax.set_ylabel('Standard Deviation -> q(x)')
ax.set_xlabel('Time Step -> k')
ax.grid(True)
```

```python
print("Fraction of Sample Paths Plot")
fig, ax=plt.subplots(1,1)
ax.plot(tArray, mean_frac, label='fraction[(q(k) <= 0)/k]')
ax.grid(True)
ax.set_ylabel('Fraction Time')
ax.set_xlabel('Time Step -> k')
plt.show()
```

```python
#dynamics
A = np.matrix([[1]])
#process noise variance
V = np.matrix([[9]])
#measurement model
H = np.matrix([[1]])
#measurement noise variance
W = np.matrix([[25]])
```

```python
measurements_z = [np.nan, 32.1 , 39.8, 45.9, 52.0, 51.0, 63.4, 58.1, 60.3, 76.6, 73.0]
inputs_uf = [10, 10, 10, 10, 10, 10, 10, 10, 10, 10, 10]
m = [7, 7, 7, 7, 7, 7, 7, 7, 7, 7, 7]
```

```python
hat_x_m = np.mat([[20]]) 
P_m = np.mat([[25]])
```

```python
numSteps = 10 
hist_hat_x_m = np.zeros([numSteps+1])
hist_P_m_diag = np.zeros([numSteps+1])  
hist_hat_x_m[0] = hat_x_m[0]
hist_P_m_diag[0] = P_m[0]
for k in np.arange(1,numSteps+1):
    uf = inputs_uf[k-1] - m[k-1]
    u = np.matrix([[uf]])
    hat_x_p = A*hat_x_m + u
    P_p = A**2*P_m + V
    z = measurements_z[k]
    K = P_p @ H.T @ np.linalg.inv(H @ P_p @ H.T + W)
    hat_x_m = hat_x_p + K @ (z - H @ hat_x_p)
    P_m = (np.eye(1) - K @ H) @ P_p @ (np.eye(1) - K @ H).T + K @ W @ K.T
    hist_hat_x_m[k] = hat_x_m[0]
    hist_P_m_diag[k] = P_m[0]
```

```python
fig, ax = plt.subplots(1,1,sharex=True, figsize=(10, 5))
ax.plot(hist_hat_x_m[:],'r.-',label="Estimated State")
ax.plot(hist_hat_x_m[:]+np.sqrt(hist_P_m_diag[:]),'b:',label="1 Sigma")
ax.plot(hist_hat_x_m[:]-np.sqrt(hist_P_m_diag[:]),'b:',)
ax.plot(measurements_z,'rx',label="Measerments z(k)")
ax.set_ylabel('r(k)')
ax.set_xlabel('Time Step -> k')
ax.legend()
ax.grid(True)
```

```python
fig, ax = plt.subplots(1,1,sharex=True, figsize=(10, 5))
ax.plot(hist_P_m_diag[:],'r.-',label="P(k)")
ax.set_xlabel('Time Step -> k')
ax.set_ylabel('P(k)')
ax.legend()
ax.grid(True)
```

```python
hat_x_m = np.mat([[20],[7]]) 
P_m = np.mat([[25,0],[0,1]])
```

```python
numSteps = 10  
hist_hat_x_m = np.zeros([numSteps+1,2])  
hist_P_m_diag = np.zeros([numSteps+1,2])  
hist_hat_x_m[0,0] = hat_x_m[0,0]
hist_hat_x_m[0,1] = hat_x_m[1,0]
hist_P_m_diag[0,0] = P_m[0,0]
hist_P_m_diag[0,1] = P_m[1,1]
for k in np.arange(1,numSteps+1):
    uf = inputs_uf[k-1]
    u = np.matrix([[uf],[0]])    
    hat_x_p = A @ hat_x_m + u
    P_p = A @ P_m @ A.T + V
    z = measurements_z[k]
    K = P_p @ H.T @ np.linalg.inv(H @ P_p @ H.T+ W)
    hat_x_m = hat_x_p + K @ (z - H @ hat_x_p)
    I = np.matrix([[1,0],[0,1]])
    P_m = (I-K @ H)@ P_p @(I-K @ H).T + K @ W @ K.T
    hist_hat_x_m[k,0] = hat_x_m[0,0]
    hist_hat_x_m[k,1] = hat_x_m[1,0]
    hist_P_m_diag[k,0] = P_m[0,0]
    hist_P_m_diag[k,1] = P_m[1,1]
```

```python
def get_A(x):
    return 1
def get_L(x):
    return -1
def get_H(x):
    return 3*(x-1)**2
def get_M(x):
    return 1
#Define noise variances
V = 0.05**2
W = 0.1**2
j0=0.1
```

```python
numSteps = 9
hat_x_m = 1
P_m = 0
hist_hat_x_m = np.zeros([numSteps+1,1])  
hist_P_m = np.zeros([numSteps+1,1]) 
hist_true_x = np.zeros([numSteps+1,1])  
true_state_x = np.random.normal(1,0)
hist_z = np.zeros([numSteps+1,1])
hist_hat_x_m[0,0] = hat_x_m
hist_P_m[0,0] = P_m
hist_true_x[0,0] = true_state_x
for k in np.arange(1,numSteps+1):
   #process noise
    v = np.random.normal(0, 0.05)
    #simulate the rocket:
    new_x = true_state_x - j0 - v
    true_state_x = new_x
    #generate the measurement:
    w = np.random.normal(0, 0.1)
    z = 4 + ((true_state_x)-1)**3 + w
    #z=measurment_z[k-1]
    
    hist_z[k,0]=z
    hat_x_p = hat_x_m - j0 - 0 
    A = get_A(hat_x_p)
    L = get_L(hat_x_p)
    P_p = A*P_m*A + L*V*L 
    
    #step 2: Kalman filter measurement update:
    H = get_H(hat_x_p)
    M = get_M(hat_x_p)
    K = P_p*H*1/(H*P_p*H + W) 
    
    predictedMeas = 4 + ((hat_x_p)-1)**3 + 0 
    hat_x_m = hat_x_p + K*(z-predictedMeas)
    I = 1
    P_m = (I - K*H)*P_p*(I - K*H) + K*W*K # (no transpose, all scalars)
    ## EKF END
    #store the variables for plotting. Don't modify the below lines:
    hist_hat_x_m[k,0] = hat_x_m
    hist_P_m[k,0] = P_m
    hist_true_x[k,0] = true_state_x
```

```python
fig, ax = plt.subplots(1,1,sharex=True, figsize=(10, 5))
ax.plot(hist_hat_x_m[:,0],'b.-',label="est")
ax.set_xlim(0,9)
ax.set_xlabel('Time k')
ax.set_ylabel('State estimated state of charge')
ax.legend()
ax.grid(True)
plt.show()
```

```python
def q(x, u, v):
    return  x - u - v  
```

```python
nSteps = 9
x0 = 1
#constants
pp=0.1
W=0.1
# Time step array
tArray = np.arange(nSteps+1) #same for all ensembles
# State history array
xArray = np.zeros(nSteps+1)
hArray = np.zeros(nSteps+1)
uArray = np.zeros(nSteps+1)
a=np.zeros(nSteps+1)
b=np.zeros(nSteps+1)
xArray[0]=x0
hArray[0]=(xArray[0]-1)**2
a[0]=hArray[0]**2*pp
b[0]=hArray[0]**2*pp+W
uArray[0]=a[0]/b[0]
for k in range(nSteps):
        u = 0.1
        v = np.random.normal(0,0.05)
        xArray[k+1] = q(xArray[k], u, v)
        hArray[k+1] = 3*(xArray[k+1]-1)**2
        a[k+1]=pp*hArray[k+1]**2
        b[k+1]=pp*hArray[k+1]**2+W
        uArray[k+1] = a[k+1]/b[k+1]
```
