```python
# importing libraries
from sympy import *
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
# Sinse:
j0 = 0.1
sigma = 0.05
# then:
K1 = ((sigma*3 + np.sqrt(sigma**2*9 +j0*4))/-j0*2)**2
K2 = ((sigma*3 -  np.sqrt(sigma**2*9 +j0*4))/-j0*2)**2

print(f'In case 1 it required {round(K1,2)} secound')
print(f'In case 2 it required {round(K2,2)} secound')
```

```python
j0 = 0.1
sigma = 0.05
K0 = 1
Ks = np.zeros([10**6,20])
times = np.arange(20)
for a in range(10**6):
    for b in range(20):
        Ks[a,b] = K0
        v = np.random.normal(0,sigma)
        K0 -= j0-v
```

```python
fig, ax = plt.subplots(3,1,sharex=True)
fig.set_size_inches(10,10)

for n,m in enumerate([np.mean(Ks,0),np.std(Ks,0),np.sum(Ks<0,0)]):

    ax[n].plot(times,m,'.-',label=f"volume of water ({n+1})")
    ax[n].grid()
```
