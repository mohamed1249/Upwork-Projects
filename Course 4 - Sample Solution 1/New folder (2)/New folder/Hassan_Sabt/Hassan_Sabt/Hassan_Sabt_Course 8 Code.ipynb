## Solution No 2

Coefficient of drag for rigid sphere is:-

$$C_d = \frac {24} {Re} + 2.6 \frac {Re/5} {1 + {(Re/5})^{1.52}} + 0.411 \frac {{(Re/263000)}^{-7.94}} {1+(Re/263000)^{-8}} + 0.25 \frac {Re x 10^{-6}} {1 + Re x 10^{-6}} - - - - (2a)$$

Given Values;

$𝜌_𝐿 = 986.15 𝐾𝑔/𝑚3$

$𝜌_𝑉 = 0.099799𝐾𝑔/𝑚3$

$𝜇_𝐿 = 0.00051198 𝑃𝑎.s$

$𝜇_𝑉 = 1.065 × 10−5 𝑃𝑎.s$

$D = 120 𝜇𝑚$

**The Maximum velocity at which the vapor can flow to not carry larger droplets (𝑈_𝑣 =? )**

### Force Balance Formula 

$$F_{Gravity} = F_{Buoyancy} + F_{Drag}$$

$$𝜌_{(liquid)} \mu (\frac{d^3} {6}) g = 𝜌_{(vapor)} \mu (\frac{d^3} {6}) + \frac 1 2 𝜌_{(liquid)} \mu (\frac{d^3} {4}) C_d (V_{vapor})^2 $$

$$C_d{(V_{vapor})^2} = \frac {4dg(𝜌_{liquid} - 𝜌_{vapor})} {3𝜌_{vapor}}$$

```python
import matplotlib.pyplot as plt
import numpy as np
from colored import fore, back, style

print (back.RED + style.BOLD + "Constant Values" + style.RESET)

problem_2_water_dia = 0.000120             
problem_2_gravity_value = 9.81                
problem_2_rho_underscore_l = 986.15           
problem_2_rho_underscore_v = 0.099799         
problem_2_mio_underscore_v = 0.00001065 
problem_2_velocity = 1

def C_underscore_D(uv):
    Re_value = (problem_2_rho_underscore_v * uv * problem_2_water_dia) / problem_2_mio_underscore_v
    return (24 / Re_value) + (2.6 * (Re_value/5) / (1 + (Re_value/5) ** 1.52)) +\
    (0.411 * (Re_value/263000) ** (-7.94) / (1 + (Re_value/263000) ** (-8))) \
    + (0.25 * (Re_value * 10**-6) / (1 + Re_value*10**-6))

def F_function(uv):
    return problem_2_velocity**2*(3*problem_2_rho_underscore_v*C_underscore_D(uv))/(4*problem_2_gravity_value*problem_2_rho_underscore_l)

# print (back.RED + style.BOLD + "Call Functions and Display Graphs" + style.RESET)
plt.figure(figsize=(10, 5))
plt.plot(np.linspace(1,2000), F_function(np.linspace(1,2000)), color='red')
plt.show()
```

```python
# print (back.RED + style.BOLD + "Bisection method" + style.RESET)
xvals = [0.2, 1.4]
t0 = F_function(xvals[0])
t1 = F_function(xvals[1])
fvals= [t0, t1]
# print (back.RED + style.BOLD + "xvals = "+str(xvals) + style.RESET)
# print (back.RED + style.BOLD + "fvals = "+str(fvals) + style.RESET)
```

```python
tolerance = 0.00005
maximum_iteration = 50
counter = 0
while True:
    x_mid_value = np.mean(xvals)
    f_mid_value = F_function(x_mid_value)
    if x_mid_value < 0:
        xvals[0], fvals[0] = x_mid_value, f_mid_value  
    elif f_mid_value > 0:
        xvals[1], fvals[1] = x_mid_value, f_mid_value
    if xvals[1]-xvals[0] < tolerance:
        break
    counter = counter + 1
    if counter > maximum_iteration:break
        
# print (back.RED + style.BOLD + "Maximum velocity at which the vapor can flow to not carry larger droplets" + style.RESET)
print('U_underscore_v =', xvals[0])
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sci

# print (back.RED + style.BOLD + "Define Constants for Problem 3" + style.RESET)
Cd_constant = 0.965 
rho_constant = 998.29
A_constant = 0.004
beta_constant = 0.5
mass_flow_rate_constant = 20

rho_tild = 0.04 * rho_constant
A_tild = 0.01 * A_constant 
beta_tild = 0.03 * beta_constant
mass_flow_rate_tild = 0.05 * mass_flow_rate_constant
b4 = beta_constant**4

# print (back.RED + style.BOLD + "Calculated Values are:-" + style.RESET)
Cd_constant, rho_constant, A_constant, beta_tild, mass_flow_rate_constant, rho_tild, A_tild, beta_tild, mass_flow_rate_tild, b4 
```

```python
value = mass_flow_rate_tild / Cd_constant / A_constant
dp = (value**2)*(1-b4) / rho_constant / 2
dp_tild = dp*np.sqrt((2*mass_flow_rate_tild / mass_flow_rate_constant)**2 + (-2 * A_tild / A_constant)**2 + \
                      (-1 * rho_tild / rho_constant)**2 + ((-4 * b4/(1-b4)) * beta_tild / beta_constant)**2)

# print (back.RED + style.BOLD + "dp =" +str(dp) + style.RESET)
# print (back.RED + style.BOLD + "dp_tild =" +str(dp_tild) + style.RESET)
```

## Monte Carlo method

```python
# print (back.RED + style.BOLD + "Monte Carlo Simulation" + style.RESET)
number_of_iterations = 1000000
dp_samples = np.empty(number_of_iterations)
for i in range(number_of_iterations):
    rho_samples, A_samples = np.random.normal(rho_constant, rho_tild / 2), np.random.normal(A_constant, A_tild / 2)
    beta_samples, mdot_samples = np.random.normal(beta_constant, beta_tild / 2), \
    np.random.normal(mass_flow_rate_constant, mass_flow_rate_tild / 2)    
    dp_samples[i] = ((mdot_samples / Cd_constant / A_samples)**2)*(1-beta_samples**4) / rho_constant / 2

# print (back.RED + style.BOLD + "The Average of DP_SAMPLES is:- " + style.RESET)
np.mean(dp_samples)
```

```python
print (back.RED + style.BOLD + "Define Constants for Problem 4" + style.RESET)

rho_constant = 998.2     
mu_constant = 0.001002   
mass_flow_rate_constant = 750
d_before_constant = 0.8
eps_before_constant = 4.5*1e-3
gravity_constant = 9.81
L_constant = 400
eta_constant = 0.82
rate_before_constant = 0.15

def costFunction(Params):
    d_before,eps_before,rate_before = Params
    v = mass_flow_rate_constant / rho_constant / (np.pi*d_before**2/4)
    Re = rho_constant * v * d_before / mu_constant
    log_values = np.log10((eps_before / d_before / 3.7)**(1.11)+(6.9/Re))
    f_value = (-1.8 * log_values)**(-2)
    h_value = f_value * (L_constant / d_before) * ((v**2)/(2 * gravity_constant))
    return (rho_constant * gravity_constant * (mass_flow_rate_constant / rho_constant) * \
            h_value / eta_constant / 1000) *(24*365)*rate_before
```

```python
I = 2300
def runMonteCarlo(wn, derr, lerr, do_eps, I):   
    Pressure_values = np.empty((I, N_junctions))
    Pump_Power = np.empty((I,2)) 
    d_mean = wn.query_link_attribute('diameter').to_numpy()
    l_mean = wn.query_link_attribute('length').to_numpy()
    eps_mean = wn.query_link_attribute('roughness').to_numpy() 
    for i in range(I):
        count = 0
        for pipe_name, pipe in wn.pipes():
            pipe.diameter = np.random.uniform((1-derr)*d_mean[count], (1+derr)*d_mean[count])
            pipe.length = np.random.uniform((1-lerr)*l_mean[count], (1+lerr)*l_mean[count])
            if do_eps:
                pipe.roughness = np.random.normal(eps_mean[count], eps_mean[count]/5)
            count = count + 1  
        # Simulate hydraulics    
        results = wntr.sim.EpanetSimulator(wn).run_sim() 
        head = results.node['head']
        demand = results.node['demand']
        Pressure_values[i,:] = pd.Series.to_numpy(head.loc[0,:]) 
        demand_list = pd.Series.to_numpy(demand.loc[0,:])
        for ii in range(2):
            Pump_Power[i,ii] = Pressure_values[i, supply_nodes[ii]]*rho*g*abs(demand_list[supply_nodes[ii]])/eta   
    return(Pressure_values, Pump_Power)
```
