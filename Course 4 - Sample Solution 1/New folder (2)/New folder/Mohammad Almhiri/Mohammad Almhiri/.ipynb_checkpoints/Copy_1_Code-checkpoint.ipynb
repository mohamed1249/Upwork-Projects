## problem 2

Vapor in a multi-effect distillation plant such as the one shown in Figure 2 vapor can carry mist droplets with salt and other contaminants. To limit the amount of contaminants that reach the vapor exit (see arrow in the figure), we may want to control the speed of the vapor so that it can carry only the liquid droplets smaller than a certain size upwards.

Question: If we have a vertical ow of saturated water vapor at 0.15 bar absolute pressure and we want to allow it to carry droplets with a maximum diameter of 120ùúáùúám, what is the maximum velocity at which the vapor can flow to not carry larger droplets?)


<center>Figure 2</center>

#### Assumptions and hints:

- Assume steam is saturated. Definition: Water heated to boiling point is vaporized and turns to saturated steam. If saturated steam is heated further (temperature above boiling point at given pressure), dry steam is generated and traces of moisture are gone (and in such case we have superheated steam).
- Assume droplet is spherical and can be treated as a rigid sphere a first approximation.
- Use the density and viscosity of liquid and vapor water at the appropriate saturation temperature. One reference for such properties can be found here: NIST - Thermophysical Properties of Fluid Systems.
- Hint: start with a force balance. Assume that droplet would need to travel vertically upward (towards the screen above).

## Solution 2

Coefficient of drag for rigid sphere can be calculated as:-

$$C_d = \frac {24} {Re} + 2.6 \frac {Re/5} {1 + {(Re/5})^{1.52}} + 0.411 \frac {{(Re/263000)}^{-7.94}} {1+(Re/263000)^{-8}} + 0.25 \frac {Re x 10^{-6}} {1 + Re x 10^{-6}} - - - - (2a)$$

Given Data

$ùúå_ùêø = 986.15 ùêæùëî/ùëö3$

$ùúå_ùëâ = 0.099799ùêæùëî/ùëö3$

$ùúá_ùêø = 0.00051198 ùëÉùëé.s$

$ùúá_ùëâ = 1.065 √ó 10‚àí5 ùëÉùëé.s$

$D = 120 ùúáùëö$

#### The Maximum velocity at which the vapor can flow to not carry larger droplets (ùëà_ùë£ =? )

### Force Balance Formula 

$$F_{Gravity} = F_{Buoyancy} + F_{Drag}$$

$$ùúå_{(liquid)} \mu (\frac{d^3} {6}) g = ùúå_{(vapor)} \mu (\frac{d^3} {6}) + \frac 1 2 ùúå_{(liquid)} \mu (\frac{d^3} {4}) C_d (V_{vapor})^2 $$

Re-arrange above equation, we get

$$C_d{(V_{vapor})^2} = \frac {4dg(ùúå_{liquid} - ùúå_{vapor})} {3ùúå_{vapor}}$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
diameter_of_water = 0.000120             
gravity = 9.81                
rho_l = 986.15           
rho_v = 0.099799         
mio_v = 0.00001065 

def function_Cd(uv):
    Re = rho_v*uv*diameter_of_water/mio_v
    cd = (24 / Re) + (2.6 * (Re/5) / (1 + (Re/5) ** 1.52)) +\
    (0.411 * (Re/263000) ** (-7.94) / (1 + (Re/263000) ** (-8))) \
    + (0.25 * (Re * 10**-6) / (1 + Re*10**-6))
    return cd

def function_f(uv):
    f_uv = (uv**2) * function_Cd(uv) - diameter_of_water * gravity * 4 * rho_l / rho_v / 3
    return f_uv

plt.figure(figsize=(12, 6))
Uv = np.linspace(1,1000)
plt.plot(Uv, function_f(Uv))
plt.grid()
```

```python
#Bisection method
tolerance = 1e-5
maxit = 30
xvals = [0.2,1.4]
fvals = [function_f(xvals[0]),function_f(xvals[1])]
xvals, fvals
```

```python
count = 0
while True:
    xmid = np.mean(xvals)
    fmid = function_f(xmid)
    if fmid < 0:
        xvals[0], fvals[0] = xmid, fmid  
    elif fmid > 0:
        xvals[1], fvals[1] = xmid, fmid
    
    if xvals[1]-xvals[0] < tolerance:
        break
    
    count = count + 1
    if count > maxit:
        break
print("Solution")
print('The  maximum velocity at which the vapor can flow to not carry larger droplets Uv =', xvals[0])
```

Our venturi has Œ≤ = 0:6 and Athroat = 0:003 m2. We shall assume that the uncertainty in density is 2 %, in area of throat 1 % and in Œ≤ is 1 % for each, all with 95 % confidence. If mass flow rate of 25 kg/s is to be determined within 5 % with 95 % confidence, what is maximum uncertainty allowable for the differential pressure measurement?

### (a) Use the Gauss method

```python
Cd = 0.965 
rho = 998.29
A = 0.003
beta = 0.6
mass_flow_rate = 25

rho_tilde = 0.02*rho
A_tilde = 0.01*A 
beta_tilde = 0.01*beta
mass_flow_rate_tilde = 0.05*mass_flow_rate
b4 = beta**4
Cd, rho, A, beta, mass_flow_rate, rho_tilde, A_tilde, beta_tilde,mass_flow_rate_tilde, b4 
```

```python
dp = ((mass_flow_rate_tilde/Cd/A)**2)*(1-b4)/rho/2

dp_tilde = dp*np.sqrt((2*mass_flow_rate_tilde/mass_flow_rate)**2 + (-2*A_tilde/A)**2 + \
                      (-rho_tilde/rho)**2 + ((-4*b4/(1-b4))*beta_tilde/beta)**2)
dp, dp_tilde
```

### (b) Use the Monte Carlo method

```python
iterations = 1000000
dp_samp = np.empty(iterations)
for i in range(iterations):
    rho_samp, A_samp = np.random.normal(rho, rho_tilde/2), np.random.normal(A, A_tilde/2)
    beta_samp, mdot_samp = np.random.normal(beta, beta_tilde/2), np.random.normal(mass_flow_rate, mass_flow_rate_tilde/2)    
    dp_samp[i] = ((mdot_samp/Cd/A_samp)**2)*(1-beta_samp**4)/rho/2
    
np.mean(dp_samp)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sci

plt.figure()
plt.hist(dp_samp, bins = 100, density = True, label='Monte Carlo') # density = True normalizes histogram with pdf
plt.title(f'Monte Carlo Mean = {np.mean(dp_samp):.2f}, std = {np.std(dp_samp)*2:.2f}')

# Overlay results from Gauss Prop
x = np.linspace(min(dp_samp),max(dp_samp),100);
norm = sci.norm.pdf(x,dp,dp_tilde/2);
plt.plot(x,norm,'r-',lw=2,label='Gaussian')
plt.legend()
```

```python
rho = 1025     
mu = 0.00109   
mass_flow_rate = 850
d_before = 0.6
eps_before = 5.5*1e-3
gravity = 9.81
L = 300
eta = 0.82
rate_before = 0.11

def cost(d_before,eps_before,rate_before):
    q = mass_flow_rate / rho
    v = q / (np.pi*d_before**2/4)
    Re = rho*v*d_before/mu
    log_val = np.log10((eps_before/d_before/3.7)**(1.11)+(6.9/Re))
    f = (-1.8 * log_val)**(-2)
    h = f*(L/d_before)*((v**2)/(2*gravity))
    P = rho*gravity*q*h/eta/1000
    return P*(24*365)*rate_before
```

```python
I = 2000
def runMonteCarlo(wn, derr, lerr, do_eps, I):   
    All_Pressure = np.empty((I, N_junctions)) #col is junction row is iteration
    Pump_Power = np.empty((I,2)) # col is pump (1,5), row is iteration  
    #Distribution mean
    d_mean = wn.query_link_attribute('diameter').to_numpy() # [m]
    l_mean = wn.query_link_attribute('length').to_numpy() # [m]
    eps_mean = wn.query_link_attribute('roughness').to_numpy() # [mm]  
    for i in range(I):
        #Get d, l, eps from distribution
        #Update wn model with new values
        count = 0
        for pipe_name, pipe in wn.pipes():  #Iterate over pipes, count used to mark index of mean values
            pipe.diameter = np.random.uniform((1-derr)*d_mean[count], (1+derr)*d_mean[count])
            pipe.length = np.random.uniform((1-lerr)*l_mean[count], (1+lerr)*l_mean[count])
            if do_eps:
                pipe.roughness = np.random.normal(eps_mean[count], eps_mean[count]/5)
            count = count + 1 #index of pipe   
        # Simulate hydraulics    
        results = wntr.sim.EpanetSimulator(wn).run_sim() 
        # Process Results - store in pressure, power arrays
        # Create vectors of head, demand
        head = results.node['head'] # [m]
        demand = results.node['demand'] # [m^3/s]
        All_Pressure[i,:] = pd.Series.to_numpy(head.loc[0,:]) 
        demand_list = pd.Series.to_numpy(demand.loc[0,:])
        # Compute power and store it in an array = head*rho*g*abs(Q)/eta
        for ii in range(2):
            Pump_Power[i,ii] = All_Pressure[i, supply_nodes[ii]]*rho*g*abs(demand_list[supply_nodes[ii]])/eta
            
    return(All_Pressure, Pump_Power)
```

```python
wn = wntr.network.WaterNetworkModel(inp_file)
All_Pressure, Pump_Power = runMonteCarlo(wn, 0.02, 0.02, True, I)
```

```python
meanPressure = [np.mean(All_Pressure[:i,]) for i in range(I)]
meanPower = [np.mean(Pump_Power[:i,:]) for i in range(I)]
plt.figure(figsize=(14,6))
plt.subplot(211)
plt.plot(range(I), meanPressure, linewidth=1)
plt.grid()
plt.subplot(212)
plt.plot(range(I),meanPower,linewidth=1)
plt.grid()
```

## Bonus task:

Name three methods you could use in this water pipe network to measure the flow rate, and name the underlying physical operating principle used by each instrument. Hint: See day 9 lecture notes.

### 1. Orifice Flow Meters.



orifice meter is a conduit and a restriction to create a pressure drop. An hour glass is a form of orifice. A nozzle, venturi or thin sharp edged orifice can be used as the flow restriction. In order to use any of these devices for measurement it is necessary to empirically calibrate them. That is, pass a known volume through the meter and note the reading in order to provide a standard for measuring other quantities. As the fluid approaches the orifice the pressure increases slightly and then drops suddenly as the orifice is passed. It continues to drop until the ‚Äúvena contracta‚Äù is reached and then gradually increases until at approximately 5 to 8 diameters downstream a maximum pressure point is reached that will Orifice be lower than the pressure upstream of the orifice.

- The decrease in pressure as the fluid passes thru the orifice is a result of the increased velocity of the gas passing thru the reduced area of the orifice.
- When the velocity decreases as the fluid leaves the orifice the pressure increases and tends to return to its original level. All of the pressure loss is not recovered because of friction and turbulence losses in the stream.
- The pressure drop across the orifice increases when the rate of flow increases. When there is no flow there is no differential. The differential pressure is proportional to the square of the velocity, it therefore follows that if all other factors remain constant, then the differential is proportional to the square of the rate of flow.

### 2.	Electromagnetic Flow Meters:



Electromagnetic flow meter, also called Magmeters, it is a contact instrument used for measuring the volumetric flowrates of any fluid which can adequately conduct electricity in closed pipelines. These instruments are highly accurate, reliable, and stable devices that are used in various high-pressure industrial processes. flow meter operates on the basis of Faraday‚Äôs Law of Induction. According to this principle, any change in the magnetic flux, linked to an electric circuit, causes an electromotive force (or voltage) to be induced in this circuit. The induced voltage is therefore directly proportional to the rate of change of magnetic flux with time. It can measure highly corrosive mediums such as concentrated acid and concentrated alkaline using PTFE lining material. By using Titanium electrodes, we can measure seawater, various chloride, and hypo-chloride fluids.

### 3.	Coriolis Mass Flow Meters



The Coriolis principle refers to the effect that a moving mass has on a body in a rotating frame of reference. The moving mass exerts an apparent force on the body, causing a deformation of the body. This force is called the Coriolis force. It does not act directly on the body, but on the motion of the body. This principle is used in Coriolis flow meters.

A Coriolis flow meter consists of two parallel tubes that are made to oscillate using a magnet. These oscillations are recorded by sensors fitted at the inlet and outlet of each tube. In a no-flow state, the oscillations are synchronized, i.e. in phase with each other, since there is no mass exerting any force on the tubes. On the other hand, any fluid or gas flowing through the tubes generates Coriolis forces, causing the tubes to twist in proportion to the mass flow rate of the medium. The magnitude of this twisting is measured by the sensors as a phase shift between the inlet and outlet of the tubes. And this phase shift is used to measure the mass flow rate of the medium. The tubes also oscillate at their own resonant frequency in proportion to the density of the medium through the tubes. This effect is used to derive the density of the medium. Knowing the mass flow rate and the density, it is then possible to calculate the volume flow rate of the medium.
