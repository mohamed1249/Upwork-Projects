Rigid Sphere's Drag's Coefficient:-

$$C_d = \frac {24} {Re} + 2.6 \frac {Re/5} {1 + {(Re/5})^{1.52}} + 0.411 \frac {{(Re/263000)}^{-7.94}} {1+(Re/263000)^{-8}} + 0.25 \frac {Re x 10^{-6}} {1 + Re x 10^{-6}}$$

$𝜌_𝐿 = 986.15 𝐾𝑔/𝑚3$

$D = 120 𝜇𝑚$

#### The Maximum velocity at which the vapor can flow to not carry larger droplets  = 𝑈_𝑣 =? 

Use Force Balance Formula to Calculate Required Value

$$F_{Gravity} = F_{Buoyancy} + F_{Drag}$$

$$𝜌_{(liquid)} \mu (\frac{d^3} {6}) g = 𝜌_{(vapor)} \mu (\frac{d^3} {6}) + \frac 1 2 𝜌_{(liquid)} \mu (\frac{d^3} {4}) C_d (V_{vapor})^2 $$

Simplify

$$C_d{(V_{vapor})^2} = \frac {4dg(𝜌_{liquid} - 𝜌_{vapor})} {3𝜌_{vapor}}$$

```python
print("Python Libraries used to this assignment")
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sci
```

```python
print("Define Constant Here")           
value_of_gravity = 9.81                
value_of_rho_w = 986.15           
value_of_rho_v = 0.099799    
value_of_mio_w = 0.00051198     
value_of_mio_v = 0.00001065
alpha = (4*value_of_rho_w*value_of_gravity)/(3*value_of_rho_v)
```

```python
def CD_Vapors_Function(dk ,rho = value_of_rho_v, v = 1, mu = value_of_mio_v):
    Re = rho*v*dk/mu
    C_d = 24/Re + 2.6*(Re/5)/(1 + (Re/5)**1.52) + 0.411*(Re/263000)**(-7.94)/(1+ (Re/263000)**(-8)) + 0.25*(Re*10**-6)/(1+ Re*10**-6)
    return C_d

def finalF(dk): return CD_Vapors_Function(dk)/alpha
```

```python
print("Define Constants here for Question 4")
value_of_rho = 1025
value_of_mu = 0.00109
value_of_mass_flow_rate = 750
value_of_d_before = 0.6
value_of_eps_before = 4.5*1e-3
value_of_gravity = 9.81
value_of_L = 400
value_of_eta = 0.82
value_of_rate_before = 0.15
```

```python
def costFunction(value_of_d_before, value_of_eps_before, value_of_rate_before):
    value_of_q = value_of_mass_flow_rate / value_of_rho
    value_of_v = value_of_q / (np.pi*value_of_d_before**2/4)
    value_of_R = value_of_rho * value_of_v * value_of_d_before / value_of_mu
    value_of_log_val = np.log10((value_of_eps_before / value_of_d_before / 3.7)**(1.11)+(6.9 / value_of_R))
    value_of_f = (-1.8 * value_of_log_val)**(-2)
    value_of_h = value_of_f * (value_of_L / value_of_d_before) * ((value_of_v**2) / (2 * value_of_gravity))
    value_of_P = value_of_rho * value_of_gravity * value_of_q * value_of_h / value_of_eta / 1000
    return value_of_P * (24 * 365) * value_of_rate_before
```

> (A) What is the electricity cost per year for this single pipeline?

```python
cost_before_one = costFunction(value_of_d_before, value_of_eps_before, value_of_rate_before)
print("Eletricity Cost/ Year is equal to ${}".format(int(round(cost_before_one))))
```

> Monte Carlo Simulation

```python
value_of_I = 200
def runMonteCarlo(wn, derr, lerr, do_eps, I):   
    
    value_of_All_Pressure = np.empty((value_of_I, value_of_N_junctions))
    value_of_Pump_Power = np.empty((value_of_I,3))
    
    value_of_d_mean = wn.query_link_attribute('diameter').to_numpy()
    value_of_l_mean = wn.query_link_attribute('length').to_numpy()
    
    value_of_eps_mean = wn.query_link_attribute('roughness').to_numpy()  
    
    for i in range(value_of_I):
        count = 0
        for pipe_name, pipe in wn.pipes():
            pipe.diameter = np.random.uniform((1-derr) * \
                                              value_of_d_mean[count], (1+derr)*value_of_d_mean[count])
            pipe.length = np.random.uniform((1-lerr) * \
                                            value_of_l_mean[count], (1+lerr)*value_of_l_mean[count])
            if do_eps:
                pipe.roughness = np.random.normal(value_of_eps_mean[count], value_of_eps_mean[count]/5)
            count = count + 1   
        value_of_results = wntr.sim.EpanetSimulator(wn).run_sim() 
        value_of_head = value_of_results.node['head'] # [m]
        value_of_demand = value_of_results.node['demand'] # [m^3/s]
        value_of_All_Pressure[i,:] = pd.Series.to_numpy(value_of_head.loc[0,:]) 
        value_of_demand_list = pd.Series.to_numpy(value_of_demand.loc[0,:])
        
        for j in range(2):
            value_of_Pump_Power[i,j] = value_of_All_Pressure[i, value_of_supply_nodes[j]] * \
            value_of_rho * value_of_g * abs(value_of_demand_list[value_of_supply_nodes[j]]) / value_of_eta
            
    return(value_of_All_Pressure, value_of_Pump_Power)
```

```python
wn = wntr.network.WaterNetworkModel(value_of_inp_file)
value_of_All_Pressure, value_of_Pump_Power = runMonteCarlo(wn, 0.02, 0.01, True, value_of_I)
```
