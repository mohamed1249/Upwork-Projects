```python
import pyomo.environ as pyo
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
plt.style.use('fivethirtyeight')
```

# Load Data

```python
plantProps = pd.read_csv('plantProps.csv').to_numpy()
print(plantProps)
plantProps = plantProps[:,1:]

loadProfile = pd.read_csv('loadProfile.csv').to_numpy()
loadProfile = loadProfile[:,1]
nHours = loadProfile.shape[0]
Time = np.arange(1,nHours+1)
```

```python
GenerationTarget = loadProfile*1.05 # Add additional 5% buffer to load profile
# GenerationTarget = loadProfile
InstalledCapacity = sum(plantProps[4,:])

if InstalledCapacity < max(GenerationTarget):
    print('Error! Maximum of load profile {:2f} MW is above installed capacity {:2f} MW!'.format(max(GenerationTarget), \
                                                                                                 installedCapacity))
```

```python
plt.figure(figsize=(10, 6))
plt.figure(1)
plt.clf()
plt.plot(Time, loadProfile)
plt.plot(Time, GenerationTarget)
plt.axhline(y=InstalledCapacity, color="black", linestyle=":")
plt.xlabel('Time (hrs)')
plt.ylabel('Load (MW)')
plt.legend(['Load Profile','Generation Target','Maximum Generation Level'],loc='best')
```

```python
FuelCost = plantProps[0,:]
OperatingCost = plantProps[1,:]

# .... add other constants ..
StartupCost = plantProps[2,:]
MinGenerationLevel = plantProps[3,:]
MaxGenerationLevel = plantProps[4,:]
RampUpLimit = plantProps[5,:]
RampDownLimit = plantProps[6,:]
MinimumUpTime = plantProps[7,:]
MinimumDownTime = plantProps[8,:]
```

# Define Optimization Problem

```python
# Initialization of optimization problem  
model = pyo.ConcreteModel()

#index sets initialization
model.K=pyo.Set(initialize=np.array([i for i in range(0,nHours)]))
model.K_1N=pyo.Set(initialize=np.array([i for i in range(1,nHours)]))

#decision variables configuration
model.coal_old = pyo.Var(model.K, domain = pyo.NonNegativeReals)
model.coal_new = pyo.Var(model.K, domain = pyo.NonNegativeReals)
model.gas = pyo.Var(model.K, domain = pyo.NonNegativeReals)
model.peaker = pyo.Var(model.K, domain = pyo.NonNegativeReals)

model.max_generation=pyo.Var(model.K, domain = pyo.NonNegativeReals)
model.total_generation=sum(model.coal_old[j]+model.coal_new[j]+model.gas[j]+model.peaker[j] for j in model.K)
```

# Define Objective Function

```python
# Define Objective function here
model.obj=pyo.Objective(expr=sum(model.coal_old[i]*FuelCost[0]+\
                                 model.coal_new[i]*FuelCost[1]+\
                                 model.gas[i]*FuelCost[2]+\
                                 model.peaker[i]*FuelCost[3] \
                                 for i in model.K))
```

# Define Constraints

```python
# all constraints must be defined here
m.constraint_1=pyo.Constraint(model.K, rule=lambda model,k: sum(model.max_generation[i] \
                                                                              for i in m.K) <= MaxGenerationLevel[0])
m.constraint_2=pyo.Constraint(model.K, rule=lambda model,k: sum(model.max_generation[i] \
                                                                              for i in m.K) <= MaxGenerationLevel[1])
m.constraint_3=pyo.Constraint(model.K, rule=lambda model,k: sum(model.max_generation[i] \
                                                                              for i in m.K) <= MaxGenerationLevel[2])
m.constraint_4=pyo.Constraint(model.K, rule=lambda model,k: sum(model.max_generation[i] \
                                                                              for i in m.K) <= MaxGenerationLevel[3])
```

# Solve Optimization

```python
pyo.SolverFactory('glpk', executable="GPLK/w64/glpsol.exe").solve(model)
```
