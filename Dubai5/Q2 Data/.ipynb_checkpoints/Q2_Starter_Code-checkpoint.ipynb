```python
import random
import numpy as np
import cv2 as cv
import os
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
import tensorflow as tf
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Conv2D, MaxPool2D, Flatten, Dropout, Dense
from tensorflow.keras.losses import categorical_crossentropy
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.optimizers import Adadelta
```

```python
# Uncomment this code if data is read from Google Drive
"""
# Mount Google Drive
# you need to configure your google account at the first time running this cell
from google.colab.patches import cv2_imshow
from google.colab import drive
drive.mount('/content/drive')
"""
```

```python
### Global Setting 

# Size of the resized images 
IMAGE_WIDTH = 50
IMAGE_HEIGHT = 50
# Size of training and test data size 
TRAIN_SET = 770
TEST_SET = 1100 - 770
# PATHS
DATA_PATH = 'CleanWaterDataset'
FOLDER_PATH = 'CleanWaterProject' 
```

```python
### Load Data Methods
def load_input(filepath):
    input_img = np.load(filepath)
    input_data = input_img.reshape((-1, IMAGE_HEIGHT, IMAGE_WIDTH, 1)) 
    input_data = input_data.astype('float32')
    input_data /= 255 
    return input_img, input_data

def load_label(filepath):
    label = np.load(filepath)
    label_categ = to_categorical(label, 11)
    return label, label_categ

def load_cleanwater_to_train(folderpath):
    train_img, train_data = load_input(os.path.join(folderpath, 'train_input.npy'))
    train_label, train_label_categ= load_label(os.path.join(folderpath, 'train_label.npy'))
    return train_img, train_data, train_label, train_label_categ

def load_cleanwater_to_test(folderpath):
    test_img, test_data = load_input(os.path.join(folderpath,'test_input.npy'))
    test_label, test_label_categ = load_label(os.path.join(folderpath,'test_label.npy'))
    return test_img, test_data, test_label, test_label_categ
```

```python
### Load Dataset
train_img, train_x, train_y_idx, train_y = load_cleanwater_to_train(DATA_PATH)
test_img, test_x, test_y_idx, test_y = load_cleanwater_to_test(DATA_PATH)
# test 
index = 0
plt.figure(figsize=(3, 2.25))
img = train_img[index, :, :].astype(int)
plt.imshow(img, cmap='gray')
plt.title('img to train, label:{}'.format(int(train_y_idx[index])))
plt.show()
plt.figure(figsize=(3, 2.25))
img = test_img[index, :, :].astype(int)
plt.imshow(img, cmap='gray')
plt.title('img to test, label:{}'.format(int(test_y_idx[index])))
plt.show()
```

```python
### Training
from tensorflow.keras.layers import Activation, MaxPooling2D

# Define A Training Model
model = Sequential()

# Create Deep Neural Network Here
model.add(Conv2D(128, (3, 3), input_shape=train_x.shape[1:], activation="relu"))
model.add(MaxPooling2D(2,2))
model.add(Conv2D(64,(3,3), activation="relu"))
model.add(MaxPooling2D(2,2))
model.add(Conv2D(32,(3,3), activation="relu"))
model.add(Flatten())
model.add(Dense(11))
model.add(Activation('softmax'))

print("Summary of Model")
model.summary()
```

```python
# I have tried adadelta optimizer but model is not improving, then i tried SGD
# Training Routine 
from tensorflow.keras.optimizers import SGD

sgd = SGD(decay=1e-7, momentum=.9)
# it is multiclass problem therefor using categorical_crossentropy
model.compile(loss='categorical_crossentropy', optimizer=sgd, metrics=['acc'])

# I tried multiple batch sizes and epochs in order to get high accuracy
print("Epochs = 1000 \nBatch Size = 256")
batch_size = 512
epochs = 1000

# model training
model.fit(train_x, train_y, batch_size=batch_size, epochs=epochs)

#save model for later reuse
model.save(os.path.join(FOLDER_PATH,'cleanwaterai.h5'))
```

```python
loss_value, accuracy_value = model.evaluate(test_x, test_y, verbose=2)
print('loss:%.4f accuracy:%.4f' % (loss_value, accuracy_value))
```

```python
### Test
for i in range(20):
    index = random.randint(0, TEST_SET)
    pred_one_hot = model.predict(np.reshape(test_x[index, :, :, :], (1, IMAGE_HEIGHT, IMAGE_WIDTH, 1)))
    pred = np.argmax(pred_one_hot)
    plt.figure(figsize=(3, 2.25))
    img = test_img[index, :, :].astype(int)
    plt.imshow(img, cmap='gray')
    plt.title('true:%s, predicted:%s' % (int(test_y_idx[index]), pred))
    plt.show()
```
