```python
from app_store_scraper import AppStore

import pandas as pd

import numpy as np

import json

def app_store_scrapper(name,id,n_reviews):
    old_scrape = pd.DataFrame(columns=['title', 'userName', 'rating', 'review', 'isEdited',
       'date']) #creating function to automate the steps
    while True:
        app = AppStore(country='us', app_name=name, app_id = id)

        app.review(how_many=9999999999999)

        df = pd.DataFrame(np.array(app.reviews),columns=['review'])

        new_scrape = df.join(pd.DataFrame(df.pop('review').tolist()))
        new_scrape.drop('developerResponse',axis=1, inplace=True)
        
        old_scrape = pd.concat([new_scrape,old_scrape]).reset_index().drop('index', axis=1)
        old_scrape.drop_duplicates(inplace=True)

        if len(old_scrape) >= n_reviews:
            old_scrape.to_csv('F:\python\projects\ovia.csv') # make sure you don't miss the directory
            old_scrape.to_json('F:\python\projects\ovia.json')
            break
    return old_scrape

app_store_scrapper('ovia-fertility-period-tracker','570244389',3000)
```
